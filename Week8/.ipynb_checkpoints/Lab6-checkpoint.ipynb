```python
import pandas as pd
import numpy as np
from sklearn.pipeline import Pipeline
from sklearn.compose import make_column_selector, ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder, PolynomialFeatures
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import r2_score
from plotnine import ggplot, aes, geom_point
from sklearn.model_selection import GridSearchCV
```

```python
df = pd.read_csv('Hitters.csv')
#even though there are 50-ish NaN values, dropping them makes more sense than making them 0 in the case of salary, since that would throw off results a lot
df = df.dropna()
```

```python
X = df.drop(['Salary'],axis=1)
y = df['Salary']
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

# Part 1: Different Model Specs

## A: Regression without regularization

```python
# Create a pipeline that includes all the columns as predictors for salary and performs ordinary linear regression
ct = ColumnTransformer(
  [
    ("dummify",
    OneHotEncoder(sparse_output = False, handle_unknown='ignore'),
    make_column_selector(dtype_include=object)),
    ("standardize",
    StandardScaler(),
    make_column_selector(dtype_include=np.number))
  ],
  remainder = "passthrough"
)

lr_pipeline_1 = Pipeline(
  [("preprocessing", ct),
  ("linear_regression", LinearRegression())]
)
```

```python
# Fit this pipeline to the full dataset, and interpret a few of the most important coefficients
lr_fitted = lr_pipeline_1.fit(X, y)

feature_names = lr_pipeline_1.named_steps['preprocessing'].get_feature_names_out()

coefficients = lr_pipeline_1.named_steps['linear_regression'].coef_

coef_df = pd.DataFrame({
    'Variable': feature_names,
    'Coefficient': coefficients
})

coef_df['absv'] = abs(coef_df['Coefficient'])
coef_df=coef_df.sort_values(by='absv', ascending=False)
coef_df.head()
```

The variables with the five largest (positive or negative) coefficients are CRuns, CAtBat, Hits, AtBat, and CRBI. This shows that the number of times at bat (career or this season) is very important to predicting salary of a player in this model, as well as the other three variables. However, the at bat number has a negative coefficient, meaning the more times they are at bat, the lower the predicted salary. This does not make a ton of sense (although I don't know a ton about baseball).

```python
# use cross validation to estimate the MSE you would expect if you used this pipeline to predict 1989 salaries
scores1=cross_val_score(lr_pipeline_1, X, y, cv = 5, scoring = 'neg_mean_squared_error')
scores1 = abs(scores1)
scores1
```

The MSE values are all very high for this original model.

## B: Ridge Regression

```python
#Create a pipeline that includes all the columns as predictors for Salary, and performs ordinary ridge regression
#use cross validation to tune lambda
ridge_pipeline = Pipeline(
  [("preprocessing", ct),
  ("ridge_regression", Ridge())]
)

lambdas = {'ridge_regression__alpha':[0.1,1,10,100, 200, 300, 400, 500]}

gscv = GridSearchCV(ridge_pipeline, lambdas, cv = 5, scoring='r2')

gscv_fitted = gscv.fit(X, y)

results = pd.DataFrame(data = {"lambdas": [0.1,1,10,100, 200, 300, 400, 500], "Ridge scores": gscv_fitted.cv_results_['mean_test_score']})
results
```

Highest score: 0.386 with a lambda value of 200

```python
#Fit the pipeline with your chosen to the full dataset, and interpret a few of the most important coefficients.
best_ridge_pipeline = Pipeline(
  [("preprocessing", ct),
  ("ridge_regression", Ridge(alpha=200))]
)

lr_fitted = best_ridge_pipeline.fit(X, y)

feature_names = best_ridge_pipeline.named_steps['preprocessing'].get_feature_names_out()

coefficients = best_ridge_pipeline.named_steps['ridge_regression'].coef_

coef_df = pd.DataFrame({
    'Variable': feature_names,
    'Coefficient': coefficients
})

coef_df['absv'] = abs(coef_df['Coefficient'])
coef_df=coef_df.sort_values(by='absv', ascending=False)
coef_df.head()
```

In the ridge regression, the coefficient values are significantly lower than those in the original regression. Hits and runs still matter a lot, but now PutOuts and CRBI are in the top 5, and at bat (career or season) is not in the top five. Also, none of the top five coefficients are negative now.

```python
# Report the MSE you would expect if you used this pipeline to predict 1989 salaries.
scores2=cross_val_score(best_ridge_pipeline, X, y, cv = 5, scoring = 'neg_mean_squared_error')
scores2 = abs(scores2)
scores2
```

These MSEs are about as bad as the MSEs for the last model

## C: Lasso Regression

```python
#Create a pipeline that includes all the columns as predictors for Salary, and performs ordinary lasso regression
#use cross validation to tune lambda

lasso_pipeline = Pipeline(
  [("preprocessing", ct),
  ("lasso_regression", Lasso())]
)

lambdas = {'lasso_regression__alpha':[0.001,0.01,0.1,1,10, 25, 50, 100]}

gscv = GridSearchCV(lasso_pipeline, lambdas, cv = 5, scoring='r2')

gscv_fitted = gscv.fit(X, y)
```

```python
results = pd.DataFrame(data = {"lambdas": [0.001,0.01,0.1,1,10, 25, 50, 100], "Lasso scores": gscv_fitted.cv_results_['mean_test_score']})
results
```

The best score was 0.375, for lambda of 25

```python
#Fit the pipeline with your chosen  to the full dataset, and interpret a few of the most important coefficients.

best_lasso_pipeline = Pipeline(
  [("preprocessing", ct),
  ("lasso_regression", Lasso(alpha=25))]
)

lr_fitted = best_lasso_pipeline.fit(X, y)

feature_names = best_lasso_pipeline.named_steps['preprocessing'].get_feature_names_out()

coefficients = best_lasso_pipeline.named_steps['lasso_regression'].coef_

coef_df = pd.DataFrame({
    'Variable': feature_names,
    'Coefficient': coefficients
})

coef_df['absv'] = abs(coef_df['Coefficient'])
coef_df=coef_df.sort_values(by='absv', ascending=False)
coef_df.head()
```

This variable importance is slighlty different than the ridge regression, although many of the same variables are in the top 5. The most notable difference is that in the ridge regression, the top 5 coefficients were all in the same range. In this lasso regression, there is a much more significant decrease between coefficient values, with CRBI having the highest coefficient value that is nearly twice as large as the third most important, CRuns.

```python
# Report the MSE you would expect if you used this pipeline to predict 1989 salaries.
scores3=cross_val_score(best_lasso_pipeline, X, y, cv = 5, scoring = 'neg_mean_squared_error')
scores3 = abs(scores3)
scores3
```

The MSEs are still not very good.

## D: Elastic Net

```python
# Create a pipeline that includes all the columns as predictors for Salary, and performs ordinary elastic net regression
#use cross validation to tune lambda and alpha
elastic_pipeline = Pipeline(
  [("preprocessing", ct),
  ("en_regression", ElasticNet())]
)

vals = {'en_regression__alpha':[0.001,0.01,0.1,1,10, 25, 50, 100],
           'en_regression__l1_ratio':[0.001,0.01,0.1,0.25,0.5,0.75]}

gscv = GridSearchCV(elastic_pipeline, vals, cv = 5, scoring='r2')

gscv_fitted = gscv.fit(X, y)
```

```python
results_df = pd.DataFrame(gscv_fitted.cv_results_)
results_df = results_df[['param_en_regression__alpha', 'param_en_regression__l1_ratio', 'mean_test_score']]

results_df.rename(columns={
    'param_en_regression__alpha': 'lambda',
    'param_en_regression__l1_ratio': 'alpha',
    'mean_test_score': 'r2'
}, inplace=True)

results_df
```

```python
max_score_index = results_df['r2'].idxmax()

best_combination = results_df.loc[max_score_index, ['lambda', 'alpha']]
best_score = results_df.loc[max_score_index, 'r2']

print(f"The combination with the highest mean test score is: lambda {best_combination['lambda']}, alpha {best_combination['alpha']}")
print(f"The highest mean test score is: {best_score}")
```

```python
# Fit the pipeline with your chosen hyperparameters to the full dataset, and interpret a few of the most important coefficients.

best_elastic_pipeline = Pipeline(
  [("preprocessing", ct),
  ("en_regression", ElasticNet(alpha = 1, l1_ratio=0.25))]
)

lr_fitted = best_elastic_pipeline.fit(X, y)

feature_names = best_elastic_pipeline.named_steps['preprocessing'].get_feature_names_out()

coefficients = best_elastic_pipeline.named_steps['en_regression'].coef_

coef_df = pd.DataFrame({
    'Variable': feature_names,
    'Coefficient': coefficients
})

coef_df['absv'] = abs(coef_df['Coefficient'])
coef_df=coef_df.sort_values(by='absv', ascending=False)
coef_df.head()
```

The coefficients are more similar to that of the ridge regression, which makes sense because alpha is relatively small.

```python
# Report the MSE you would expect if you used this pipeline to predict 1989 salaries.

scores4=cross_val_score(best_elastic_pipeline, X, y, cv = 5, scoring = 'neg_mean_squared_error')
scores4 = abs(scores4)
scores4
```

The MSE values are still very high.

# Part 2: Variable Selection

```python
feature_names = best_elastic_pipeline.named_steps['preprocessing'].get_feature_names_out()

Ecoefficients = best_elastic_pipeline.named_steps['en_regression'].coef_
Lcoefficients = best_lasso_pipeline.named_steps['lasso_regression'].coef_
Rcoefficients = best_ridge_pipeline.named_steps['ridge_regression'].coef_
Ocoefficients = lr_pipeline_1.named_steps['linear_regression'].coef_

coef_df = pd.DataFrame({
    'Variable': feature_names,
    'Elastic Coefficient': Ecoefficients,
    'Lasso Coefficient': Lcoefficients,
    'Ridge Coefficient': Rcoefficients,
    'Linear Coefficient': Ocoefficients
})

#take the absolute value so that we determine the most important just by weight
coef_df_absolute = coef_df.loc[:, coef_df.columns != 'Variable'].abs()

# Add the 'Variable' column back to the dataframe
coef_df_absolute['Variable'] = coef_df['Variable']

# Reorder the columns to match the original dataframe
coef_df_absolute = coef_df_absolute[['Variable'] + [col for col in coef_df_absolute.columns if col != 'Variable']]

coef_df_absolute.sort_values(by='Elastic Coefficient', ascending=False).head(10)

```

Based on the above information...
*  The most important numeric variable is PutOuts
*  The five most important numeric variables are PutOuts, CRBI, Hits, CRuns, and CHits
*  The most important categorical variable is Division

## Using the five best numeric variables AND interactions with the one best categorical var

```python
df = pd.read_csv('Hitters.csv')
df = df.dropna()

X = df[['Hits', 'PutOuts', 'CRuns', 'CRBI', 'CHits', 'Division']]
y = df['Salary']
```

```python
ct1 = ColumnTransformer(
  [
    ("dummify",
    OneHotEncoder(sparse_output = False, handle_unknown='ignore'),
    make_column_selector(dtype_include=object)),
    ("standardize",
    StandardScaler(),
    make_column_selector(dtype_include=np.number))
  ],
  remainder = "passthrough"
).set_output(transform = "pandas")

ct2 = ColumnTransformer(
    [
        ('interaction1',PolynomialFeatures(interaction_only=True), ['standardize__Hits','dummify__Division_E']),
        ('interaction2',PolynomialFeatures(interaction_only=True), ['standardize__PutOuts','dummify__Division_E']),
        ('interaction3',PolynomialFeatures(interaction_only=True), ['standardize__CRuns','dummify__Division_E']),
        ('interaction4',PolynomialFeatures(interaction_only=True), ['standardize__CRBI','dummify__Division_E']),
        ('interaction5',PolynomialFeatures(interaction_only=True), ['standardize__CHits','dummify__Division_E']),
        ]
).set_output(transform = "pandas")
```
