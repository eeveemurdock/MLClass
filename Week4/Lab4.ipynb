```python
from plotnine import ggplot, aes, geom_bar, labs, theme, element_text, scale_y_continuous, expand_limits, geom_point, geom_abline, facet_wrap, geom_smooth, scale_x_continuous
```

```python
melted_df = merged_master.melt(id_vars=['state', 'population', 'region'],
                               value_vars=['Starbucks', 'DunkinDonuts', 'Peets', 'TimHortons', 'PaneraBread', 'CaribouCoffee', 'AuBonPain', 'CoffeeBean', 'McDonalds'],
                               var_name='chain',
                               value_name='store_count')

melted_df['store_count'] = pd.to_numeric(melted_df['store_count'], errors='coerce')

total_stores_per_state = melted_df.groupby('state')['store_count'].sum().reset_index()
top_states = total_stores_per_state.sort_values(by='store_count', ascending=False).head(6)['state'].tolist()

top_states_df = melted_df[melted_df['state'].isin(top_states)]

state_plot = (ggplot(top_states_df, aes(x='state', y='store_count', fill='chain'))
              + geom_bar(stat='identity', position='stack')
              + labs(title="Number of stores by chain in the top 6 states")
              + theme(axis_text_y=element_text(size=10))
              + expand_limits(y=0)
             )

print(state_plot)
```

Above are the six states with the highest total sum of these chains. Some interesting trends to note include the fact that Tim Hortons is only in NY and Peets is only in CA. Dunkin donuts is far more prevalent in the east coast states, like MA and NY. Starbucks and McDonalds are visibly prevalent in all six states.

```python
dunkin_counts = melted_df[melted_df['chain'] == 'DunkinDonuts'].groupby('state')['store_count'].sum().reset_index()
dunkin_counts.rename(columns={'store_count': 'dunkin_count'}, inplace=True)

state_dunkin_percentage = pd.merge(total_stores_per_state, dunkin_counts, on='state')
state_dunkin_percentage['dunkin_percentage'] = state_dunkin_percentage['dunkin_count'] / state_dunkin_percentage['store_count'] * 100

top_states_dunkin = state_dunkin_percentage.sort_values(by='dunkin_percentage', ascending=False).head(6)['state'].tolist()

top_states_df = melted_df[melted_df['state'].isin(top_states_dunkin)]

dunkinplot = (ggplot(top_states_df, aes(x='state', y='store_count', fill='chain'))
              + geom_bar(stat='identity', position='stack')
              + labs(title="Coffee Distribution for Popular Dunkin States")
              + theme(axis_text_y=element_text(size=10))
              + expand_limits(y=0)
             )

print(dunkinplot)
```

These states have the highest percentage of Dunkin Donuts. It makes sense that they are all in New England, as Dunkin Donuts is known as a New England staple. Massachusetts has SO many Dunkin Donuts!

```python
total_stores_per_region = melted_df.groupby('region')['store_count'].sum().reset_index()
top_regions = total_stores_per_region.sort_values(by='store_count', ascending=False).head(6)['region'].tolist()

top_regions_df = melted_df[melted_df['region'].isin(top_regions)]

region_plot = (ggplot(top_regions_df, aes(x='region', y='store_count', fill='chain'))
              + geom_bar(stat='identity', position='stack')
              + labs(title="Number of stores by chain by region")
              + theme(axis_text_y=element_text(size=10))
              + expand_limits(y=0)
             )

print(region_plot)
```

By region, the Northeast has the most Dunkin Donuts, as shown in the Dunkin Donuts graph. Peets only exists in the West, and Tim Hortons only exists in the Midwest and the Northeast. The South has the most amount of McDonalds.

```python
scatter_plot = (
    ggplot(melted_df, aes(x='population', y='store_count', color='chain'))
    + geom_point(alpha=0.6, size=2)
    + geom_smooth(method='lm', color='black', se=False)
    + labs(title="Store Count vs. Population",
           x="Population (in 10,000,000s)",
           y="Store Count")
    + theme(axis_text_x=element_text(size=10, hjust=1),
            axis_text_y=element_text(size=10))
    + facet_wrap('~ chain', scales="free")
    + scale_x_continuous(
        breaks=[10000000, 20000000, 30000000, 40000000],
        labels=['1', '2', '3', '4']
    )
)

print(scatter_plot)
```

The more of a 45 degree the line of best fit is, the more of a direct relationship between state population and store count. McDonald's, Panera, and Starbucks, the most popular restaurants, have the most linear relationship. On the other hand, Caribou Coffee and Tim Hortons, the least popular restaurants, do not have a strong correlation between population and store count.

## Automation

```python
def locations(url):

  """
  Takes in a menuism url and returns a clean, organized dataset that has three columns: state abbreviation, location count, company name

  Parameter
  ---------
  url : str
    A url, as a string, of a menuism restaurants by state website

  Returns
  -------
  df
    A pandas dataframe with columns of state abbreviation, location count, and company name
  """


  scrape = requests.get(url)
  soup = BeautifulSoup(scrape.content, "html.parser")

  search = soup.find_all("h1", style="text-align:left; color: #6A4A3C; text-transform: none;")
  item = search[0].text
  name = re.findall(r'([^\n\t]+) Locations Near Me', item)[0]

  rows = []
  locs = soup.find_all("a", class_="")
  for place in locs:
    text = place.text
    state = re.findall(r".+(?= {restaurant} locations )".format(restaurant=name), text)
    count = re.findall(r"(?<=locations ).+", text)


    if state and count:
        state = state[0]
        count = count[0].replace("(", "").replace(")", "").replace(",", "")

        rows.append({
                    'state':state,
                    'count':count,
                    'restaurant':name
                })

  df = pd.DataFrame(rows)

  state_abbreviations = {
    'Alabama': 'AL','Alaska': 'AK','Arizona': 'AZ','Arkansas': 'AR',
    'California': 'CA','Colorado': 'CO','Connecticut': 'CT','Delaware': 'DE',
    'Florida': 'FL','Georgia': 'GA','Hawaii': 'HI','Idaho': 'ID',
    'Illinois': 'IL','Indiana': 'IN','Iowa': 'IA','Kansas': 'KS','Kentucky': 'KY',
    'Louisiana': 'LA','Maine': 'ME','Maryland': 'MD','Massachusetts': 'MA','Michigan': 'MI',
    'Minnesota': 'MN','Mississippi': 'MS','Missouri': 'MO','Montana': 'MT','Nebraska': 'NE',
    'Nevada': 'NV','New Hampshire': 'NH','New Jersey': 'NJ','New Mexico': 'NM',
    'New York': 'NY','North Carolina': 'NC','North Dakota': 'ND','Ohio': 'OH',
    'Oklahoma': 'OK','Oregon': 'OR','Pennsylvania': 'PA','Rhode Island': 'RI','South Carolina': 'SC',
    'South Dakota': 'SD','Tennessee': 'TN','Texas': 'TX','Utah': 'UT',
    'Vermont': 'VT','Virginia': 'VA','Washington': 'WA','West Virginia': 'WV','Wisconsin': 'WI',
    'Wyoming': 'WY','District of Columbia': 'DC'
    }

  mask = df['state'].isin(state_abbreviations.keys())
  df = df[mask]

  def stateabb(state):
    return state_abbreviations[state]

  df['state']=df.apply(lambda x: stateabb(x['state']), axis=1)
  return(df)
```
