# Part 2

```python
X = pd.get_dummies(df[['sex','age','bmi','smoker','region']], drop_first = True)
y = df['charges']
```

### Problem 1

```python
X1 = X[['age']]
X_train, X_test, y_train, y_test = train_test_split(X1,y)
lr = LinearRegression()

my_pipeline1 = Pipeline(
    [('ols',lr)]
)
fitted_pipeline1=my_pipeline1.fit(X_train, y_train)
y_preds1=fitted_pipeline1.predict(X_test)

r1 = r2_score(y_test, y_preds1)
mse1 = mean_squared_error(y_test, y_preds1)

coefficients = fitted_pipeline1.named_steps['ols'].coef_

print("Coefficient Estimates:")
print("Intercept:", fitted_pipeline1.named_steps['ols'].intercept_)
print("Age:", coefficients[0])
print("R squared:")
print(r1)
print("MSE:")
print(mse1)
```

The R squared for this model is very low, which makes sense looking at my graph between age and charges - the points are very widely distributed and therefore a simple linear model would not perform well. The coefficients imply that the average charges tarts at \$3956, and for every added year you would expect a \$233 increase in charges.

### Problem 2

```python
enc = OneHotEncoder()
lr = LinearRegression()

X1 = X[['sex_male','age']]

my_new_pipeline = Pipeline(
    [('ols',lr)]
)

X_train, X_test, y_train, y_test = train_test_split(X1,y, test_size=0.25)

fitted_pipeline2=my_new_pipeline.fit(X_train, y_train)
y_preds=fitted_pipeline2.predict(X_test)
r2=r2_score(y_test, y_preds)
mse2 = mean_squared_error(y_test, y_preds)

coefficients = fitted_pipeline2.named_steps['ols'].coef_

print("Coefficients: ")
print(coefficients)
print("R squared: ")
print(r2)
print("MSE:")
print(mse2)
```

The first coefficient is for sex_male and the last one is for age. The effect of the variables varies greatly every time I run the model and the R squared is always very low. The last time I ran it, the coefficient for sex_male was 722.35 which means that the charges for men are \$722 higher. For age it was 233, which means that every year someone ages their charges increase \$233, on estimate.

The R squared and MSE are again very similar, and not very good.

Model 5 has the smallest MSE, at 21954492. Based on this, it is the best model to use.

```python
residuals = ynew - y_preds
plot_df = pd.DataFrame({'Predicted': y_preds, 'Residuals': residuals})

(
    ggplot(plot_df, aes(x='Predicted', y='Residuals'))
    + geom_point(alpha=0.5)
    + geom_hline(yintercept=0, color="red", linetype="dashed")
    + theme_minimal()
    + ggtitle("Residuals vs. Predicted Values for Model 5")
).draw()
```
