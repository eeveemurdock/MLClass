## Question 1

A minimum trail elevation gain (in meters)
A maximum trail elevation gain (in meters)
A minimum trail rating
The type of trail

```python
df['route_type'].unique()
```

```python
def recommend_trails(dataset, park=None, min_elev =None, max_elev=None, min_rating=None, trail_type=None):

    """
      Recommends a trail to a user

      Parameter
      ---------
          dataset 
            A dataframe of trail data
        min_elev: numeric
            minimum elevation
        max_elev: numeric
            maximum elevation
        min_rating: numeric
            minimum rating
        trail_type: str
            desired trail type
        

      Returns
      -------
      recs: str
        A list of recommendations
    """

    if min_elev is None:
        min_elev = 0
    if max_elev is None:
        max_elev = dataset['elevation_gain'].max()
    if min_rating is None:
        min_rating = 0
    if park is None:
        parkdf = dataset
    else:
        parkfilter = dataset['area_name'] == park
        parkdf = dataset[parkfilter]
    
    elevfilter = (parkdf['elevation_gain']>min_elev) & (parkdf['elevation_gain']<max_elev)
    eparkdf = parkdf[elevfilter]
    
    ratingfilter = (eparkdf['avg_rating']>min_rating)
    erparkdf = eparkdf[ratingfilter]
    
    if trail_type is None:
        final = erparkdf
    else:
        typefilter = erparkdf['route_type'] == trail_type
        final = erparkdf[typefilter]
    
    recs = final['name'].tolist()
  
    return recs
```

```python
np_trails = df
```

```python
recommend_trails(np_trails, "Haleakala National Park", min_elev = 1000, min_rating = 4)
```

```python
import requests
from bs4 import BeautifulSoup
```

```python
def scrape_month_temp(month = None):

    if month is None:
        my_url = "https://www.extremeweatherwatch.com/us-state-averages"
        response = requests.get(my_url)
        soup = BeautifulSoup(response.content, "html.parser")

        month = 'average'
        
    else:
        url = "https://www.extremeweatherwatch.com/us-state-averages/month-{mth}"
        my_url=url.format(mth=month)
        response = requests.get(my_url)
        soup = BeautifulSoup(response.content, "html.parser")
        
    tables = soup.find_all("table")
    table = tables[0]

    rows = []

    for i in table.find_all("tr")[1:]:
        cells = i.find_all("td")

        state_tag = cells[0].find("a") or cells[0]
        state = state_tag.text

        high_tag = cells[1].find("td") or cells[1]
        high = high_tag.text

        low_tag = cells[2].find("td") or cells[2]
        low = low_tag.text

        rows.append({
            "Month": month,
            "State":state,
            "High":high,
            "Low":low
        })

    data = pd.DataFrame(rows)
    
  ## More of your code here
  
  
    return data
```

```python
scrape_month_temp(month = "january")
```

```python
recs=recommend_trails(df, park=None, min_elev =1000, max_elev=3000, min_rating=4.5, trail_type='loop')
recs
```

```python
months = ['may', 'june', 'july', 'august']

dfs = list(map(lambda month: scrape_month_temp(month), months))

summer = pd.concat(dfs)
summer
```

```python
allinfo = summer.merge(df, left_on=('State'),right_on=('state_name'), how='inner')
allinfo.head()
```

```python
chosenhikesfilter = allinfo['name'].apply(lambda x: x in recs)
hikes=allinfo[chosenhikesfilter]
hikes.head()
```

```python
def elevtemp(el, temp):
    newtemp = float(temp)
    
    increments = int(el // 1000)
    
    newtemp -= increments * 5.4

    return newtemp
```

```python
elevtemp(1188.7200,82.6)
```

```python
hikes['High'] = hikes.apply(lambda x: elevtemp(x['elevation_gain'], x['High']), axis=1)
```

```python
hikes['Low'] = hikes.apply(lambda x: elevtemp(x['elevation_gain'],
                                              x['Low']),axis=1)
```
